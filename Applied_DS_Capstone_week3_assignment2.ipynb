 ### 1. import packages

```python
import pandas as pd
```

### 1. Read previously stored dataframe from assignment 1


```python
df_Toronto_city1 = pd.read_csv("df_Toronto_city.csv")
print('df_Toronto_city1.shape:',df_Toronto_city1.shape)
df_Toronto_city1.head(10) 
```

### 2. Read .csv file  that has the geographical coordinates of each postal code

```python
df_geo_coord = pd.read_csv("Geospatial_Coordinates.csv") 
print('df_geo_coord .shape:',df_geo_coord .shape)
df_geo_coord.head()
```

#### 2.1 Replace the column name and make identical to first column name of df_NY_city1 table

```python
df_geo_coord=df_geo_coord.rename(columns={"Postal Code": "Postcode"})

df_geo_coord.head()
```

### 3. Merge both dataframes based identical values of 'Postcode' column

```python
merged_df = df_geo_coord.merge(df_Toronto_city1, how = 'inner', on = ['Postcode'])
print('merged_df .shape:',merged_df.shape)
merged_df.head(10)
```

### 4. Re-arrange column names of merged dataframe to match with expected output

```python
merged_df = merged_df[['Postcode','Borough','Neighbourhood','Latitude','Longitude']]
print('merged_df .shape:',merged_df.shape)
merged_df.head(10)
```

### 7. Save Your output dataset for next assignment and further use
