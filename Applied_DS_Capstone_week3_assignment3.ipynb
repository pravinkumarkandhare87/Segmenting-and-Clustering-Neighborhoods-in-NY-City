### 2. Create a map of Toronto with neighborhoods superimposed on top

```python

address = 'Toronto, TO'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, MERGED_Toronto_city in zip(MERGED_Toronto_city['Latitude'], MERGED_Toronto_city['Longitude'], MERGED_Toronto_city['Borough'], MERGED_Toronto_city['Neighbourhood']):
    label = '{}, {}'.format(MERGED_Toronto_city, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
